```python
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
%config InlineBackend.close_figures=False
```

```python
point_num = 3
background = 3000
colors = list(matplotlib.colors.BASE_COLORS.keys())
```

```python
for i in range(10):
    fig, ax = plt.subplots()
    main_points = np.random.random((point_num,2))
    points = np.random.random((background,2))
    
    for (a, b) in points:
        d = 999
        c = None
        for i, (x, y) in enumerate(main_points):
            if max(abs(x-a), abs(y-b)) < d:
                d = max(abs(x-a), abs(y-b))
                c = colors[i%8]
        ax.scatter(a, b, c=c)
        
    for (x, y) in main_points:
        ax.scatter(x, y, c="#000000", s=200)
    plt.axis('equal')
    plt.show()
```
