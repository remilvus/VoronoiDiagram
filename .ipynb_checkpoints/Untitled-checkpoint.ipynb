```python
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
%config InlineBackend.close_figures=True
```

```python
point_num = 3
background = 3000
colors = list(matplotlib.colors.BASE_COLORS.keys())
```

```python
for i in range(5):
    fig, ax = plt.subplots()
    main_points = np.random.random((point_num,2))
    points = np.random.random((background,2))
    
    for (a, b) in points:
        d = 999
        c = None
        for i, (x, y) in enumerate(main_points):
            #metric = max(abs(x-a), abs(y-b))
            metric = abs(x-a) + abs(y-b)
            if metric < d:
                d = metric
                c = colors[i%8]
        ax.scatter(a, b, c=c)
        
    for (x, y) in main_points:
        ax.scatter(x, y, c="#000000", s=200)
    plt.axis('equal')
    plt.show()
```

```python
point_num = 4
fig, ax = plt.subplots()
main_points = np.random.random((point_num,2))
main_points = main_points[np.argsort(main_points[:,1]), :]
main_points = np.flip(main_points, axis=0)
for (x, y) in main_points:
    ax.scatter(x, y, c="#000000", s=200)
```

```python
def v_at_point(space, x, y):
    y_l = -space + y + x
    left = y_l > y
    y_l = y_l[left]
    y_r = space + y - x
    right = np.logical_not(left)
    y_r = y_r[right]
    return np.concatenate((y_l, y_r))

space = np.linspace(0, 1, 100)
fig, axes = plt.subplots(point_num, 1, figsize=(10,30))
for i, (x, y) in enumerate(main_points):
    ax = axes[i]
    for (x_, y_) in main_points:
        ax.scatter(x_, y_, c="#000000", s=200)
    ax.scatter(x, y, c="red", s=200)

    for (x, y) in main_points[:(i+1), :]:
        y_space = v_at_point(space, x, y)
        idx = np.logical_and(0 < y_space, y_space < 1)
        ax.plot(space[idx], y_space[idx])
        if i > 0:
            d = 999
            for j in range(i):
                if max(abs(x2 - main_points[j,0]), abs(y2 - main_points[j,1])) < d:
                    x2, y2 = main_points[j]
                    d = max(abs(x2 - main_points[j,0]), abs(y2 - main_points[j,1]))
            side = max(abs(x-x2), abs(y-y2))
            square1 = matplotlib.patches.Rectangle((x - side/2, y-side/2), side, side, fill=False)
            square2 = matplotlib.patches.Rectangle((x2 - side/2, y2-side/2), side, side, fill=False)
            ax.add_patch(square1)
            ax.add_patch(square2)
    plt.axis('equal')
   # plt.show()
plt.show()
```
